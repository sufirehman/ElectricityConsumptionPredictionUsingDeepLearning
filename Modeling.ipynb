```python
import pandas as pd
import numpy as np
from prophet import Prophet
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
# Preprocess the data for Prophet (handle duplicates and resample carefully)
df_prophet = df_selected_households[['DateTime', 'KWH/hh (per half hour) ']].rename(
    columns={'DateTime': 'ds', 'KWH/hh (per half hour) ': 'y'}
)
df_prophet['ds'] = pd.to_datetime(df_prophet['ds'])

# Group duplicates properly by averaging their 'y' values
df_prophet = df_prophet.groupby('ds').agg({'y': 'mean'}).reset_index()

# Ensure regular intervals by resampling and filling gaps
df_prophet = df_prophet.set_index('ds').asfreq('30T', method='pad').reset_index()

# Split dataset (70% train and 30% test)
train_size = int(len(df_prophet) * 0.7)
train_data = df_prophet.iloc[:train_size]
test_data = df_prophet.iloc[train_size:]

# Initialize and fit Prophet model
model_prophet = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=True,
    changepoint_prior_scale=0.005,  # Adjust this as needed
    seasonality_prior_scale=30   # Test larger values if required
)
model_prophet.fit(train_data)

# Make predictions on the test period
future = pd.DataFrame({'ds': test_data['ds']})
forecast = model_prophet.predict(future)

# Extract and align predictions with actual data
predicted = forecast[['ds', 'yhat']].set_index('ds')
actual = test_data.set_index('ds')
aligned_actual = actual.reindex(predicted.index).dropna()
aligned_predicted = predicted.loc[aligned_actual.index]


# Calculate and print metrics
rmse = np.sqrt(mean_squared_error(aligned_actual['y'], aligned_predicted['yhat']))
mae = mean_absolute_error(aligned_actual['y'], aligned_predicted['yhat'])

print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")


# Apply a simple moving average to both actual and predicted values
window_size = 5  # You can adjust this for smoother results (larger window size smooths more)

# For actual values (you can choose any method, such as rolling mean)
smoothed_actual = aligned_actual['y'].rolling(window=window_size).mean()

# For predicted values
smoothed_predicted = aligned_predicted['yhat'].rolling(window=window_size).mean()

# Plot the smoothed values
plt.figure(figsize=(12, 6))
plt.plot(smoothed_actual, label='Smoothed Actual', color='blue')
plt.plot(smoothed_predicted, label='Smoothed Predicted', color='red')
plt.title('Smoothed Facebook Prophet Forecast vs Actual')
plt.xlabel('Datetime')
plt.ylabel('Electricity Consumption (KWH)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Transform the data into sequences to allow the model to learn temporal dependencies from previous data points
import numpy as np
import pandas as pd

# Data was already normalized by min max scaler

# Split into train and test sets (70% train, 30% test)
train_size = int(len(df_selected_households) * 0.7)
train_data = df_selected_households[:train_size]
test_data = df_selected_households[train_size:]

# Convert to time-series sequences
def create_dataset(data, time_step=30):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])  # Input sequence (time_step)
        y.append(data[i + time_step, 0])  # Output (target)
    return np.array(X), np.array(y)

time_step = 30  
train_data_scaled = train_data['KWH/hh (per half hour) '].values.reshape(-1, 1)
test_data_scaled = test_data['KWH/hh (per half hour) '].values.reshape(-1, 1)

# Creating training and testing datasets
X_train, y_train = create_dataset(train_data_scaled, time_step)
X_test, y_test = create_dataset(test_data_scaled, time_step)

# Reshaping input for LSTM [samples, time_steps, features]
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
import optuna
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_squared_error
from tensorflow.keras.callbacks import EarlyStopping

# Define the objective function for Optuna
def objective(trial):
    # Define hyperparameters to tune
    lstm_units_1 = trial.suggest_int("lstm_units_1", 32, 128)
    lstm_units_2 = trial.suggest_int("lstm_units_2", 32, 128)
    dropout_rate = trial.suggest_uniform("dropout_rate", 0.1, 0.5)
    learning_rate = trial.suggest_loguniform("learning_rate", 1e-5, 1e-2)
    batch_size = trial.suggest_categorical("batch_size", [16, 32, 64, 128])
    epochs = trial.suggest_int("epochs", 10, 50)

    # Initialize the LSTM model
    model = Sequential()
    model.add(LSTM(units=lstm_units_1, return_sequences=True, input_shape=(X_train.shape[1], 1)))
    model.add(Dropout(dropout_rate))
    model.add(LSTM(units=lstm_units_2, return_sequences=False))
    model.add(Dropout(dropout_rate))
    model.add(Dense(units=1))  # Output layer

    # Compile the model
    model.compile(optimizer=Adam(learning_rate=learning_rate), loss="mean_squared_error")

    # Define early stopping to prevent overfitting
    early_stopping = EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True)

    # Train the model
    history = model.fit(
        X_train, y_train,
        epochs=epochs,
        batch_size=batch_size,
        validation_data=(X_test, y_test),
        verbose=0,
        callbacks=[early_stopping]
    )

    # Predict on the test set
    y_pred = model.predict(X_test)

    # Calculate RMSE (return this as the value to minimize)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    return rmse

# Run Optuna study to optimize hyperparameters
study = optuna.create_study(direction="minimize")  # Minimize RMSE
study.optimize(objective, n_trials=50)

# Print best parameters found by Optuna
print(f"Best hyperparameters: {study.best_trial.params}")

# Use the best parameters to train the final model
best_params = study.best_trial.params
final_model = Sequential()
final_model.add(LSTM(units=best_params["lstm_units_1"], return_sequences=True, input_shape=(X_train.shape[1], 1)))
final_model.add(Dropout(best_params["dropout_rate"]))
final_model.add(LSTM(units=best_params["lstm_units_2"], return_sequences=False))
final_model.add(Dropout(best_params["dropout_rate"]))
final_model.add(Dense(units=1))  # Output layer

# Compile with the best learning rate
final_model.compile(optimizer=Adam(learning_rate=best_params["learning_rate"]), loss="mean_squared_error")

# Train the final model
final_model.fit(
    X_train, y_train,
    epochs=best_params["epochs"],
    batch_size=best_params["batch_size"],
    validation_data=(X_test, y_test),
    verbose=1
)

# Make final predictions
y_pred_final = final_model.predict(X_test)

# Calculate final RMSE and MAE
rmse_final = np.sqrt(mean_squared_error(y_test, y_pred_final))
mae_final = mean_absolute_error(y_test, y_pred_final)

print(f"Final RMSE: {rmse_final:.4f}")
print(f"Final MAE: {mae_final:.4f}")
```

 ### Seq2Seq Model

```python
# Seq2Seq Model which will help us forecast for the n steps ahead
# Compared to LSTM it will forecast till far and works on encoder decoder structure

from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense
import numpy as np

# Define input and output sequences
time_step = 30
forecast_steps = 5  # Number of steps to predict ahead

# Prepare input-output pairs
def create_seq2seq_dataset(data, time_step, forecast_steps):
    X, y = [], []
    for i in range(len(data) - time_step - forecast_steps):
        X.append(data[i:i+time_step, 0])  # Input sequence
        y.append(data[i+time_step:i+time_step+forecast_steps, 0])  # Output sequence
    return np.array(X), np.array(y)

# Creating train and test datasets
X_train_seq, y_train_seq = create_seq2seq_dataset(train_data_scaled, time_step, forecast_steps)
X_test_seq, y_test_seq = create_seq2seq_dataset(test_data_scaled, time_step, forecast_steps)

# Reshaping for LSTM input format [samples, timesteps, features]
X_train_seq = X_train_seq.reshape(X_train_seq.shape[0], X_train_seq.shape[1], 1)
X_test_seq = X_test_seq.reshape(X_test_seq.shape[0], X_test_seq.shape[1], 1)

# Seq2Seq Model Architecture
encoder_input = Input(shape=(time_step, 1))  # Input for encoder
encoder_lstm = LSTM(50, return_state=True)
encoder_output, state_h, state_c = encoder_lstm(encoder_input)
encoder_states = [state_h, state_c]  # Encoder state is passed to the decoder

decoder_input = Input(shape=(forecast_steps, 1))  # Input for decoder
decoder_lstm = LSTM(50, return_sequences=True, return_state=True)
decoder_output, _, _ = decoder_lstm(decoder_input, initial_state=encoder_states)
decoder_dense = Dense(1)  # Output layer to make predictions
decoder_output = decoder_dense(decoder_output)

# Combining the encoder and decoder into a single model
seq2seq_model = Model([encoder_input, decoder_input], decoder_output)

# Compile the model
seq2seq_model.compile(optimizer='adam', loss='mean_squared_error')

# Train the Seq2Seq model
# Since for Seq2Seq, decoder_input should be given separately, we'll initialize it with zeros or previously predicted outputs
decoder_input_train = np.zeros((X_train_seq.shape[0], forecast_steps, 1))  # initializing decoder input with zeros
decoder_input_test = np.zeros((X_test_seq.shape[0], forecast_steps, 1))  # similarly for testing
```

```python
# Lets train the model

seq2seq_model.fit([X_train_seq, decoder_input_train], y_train_seq, epochs=20, batch_size=64, validation_data=([X_test_seq, decoder_input_test], y_test_seq))
```

```python
# Predict output sequences for the test data
decoder_input_pred = np.zeros((X_test_seq.shape[0], forecast_steps, 1))
y_pred_seq2seq = seq2seq_model.predict([X_test_seq, decoder_input_pred])

# Select the first value of each forecast sequence for comparison with actuals
# (Adjust this step as needed based on your specific comparison goals)
y_pred_seq2seq_first = y_pred_seq2seq[:, 0].flatten()
y_test_first = y_test_seq[:, 0].flatten()

# Adjust the test dates to match predictions
test_dates = test_data['DateTime'].iloc[time_step + forecast_steps: time_step + forecast_steps + len(y_test_first)].values

# Check for length consistency before proceeding to the plot
assert len(test_dates) == len(y_test_first) == len(y_pred_seq2seq_first), \
    f"Mismatch in lengths: test_dates({len(test_dates)}), y_test({len(y_test_first)}), predictions({len(y_pred_seq2seq_first)})"

# Plot actual vs. predicted values
plt.figure(figsize=(14, 6))
plt.plot(test_dates, y_test_first, color='blue', label='Actual Consumption')
plt.plot(test_dates, y_pred_seq2seq_first, color='red', label='Predicted Consumption')

# Format the plot
plt.title('Actual vs Predicted Electricity Consumption (Seq2Seq)')
plt.xlabel('DateTime')
plt.ylabel('Electricity Consumption (scaled)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

#### Optuna for Seq2Seq

```python
import optuna
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense
import tensorflow as tf
import matplotlib.pyplot as plt

# Function to create and return a Seq2Seq model with hyperparameters
def create_seq2seq_model(time_step, forecast_steps, lstm_units, dropout_rate, learning_rate):
    encoder_input = Input(shape=(time_step, 1))  # Input for encoder
    encoder_lstm = LSTM(lstm_units, return_state=True, dropout=dropout_rate)
    encoder_output, state_h, state_c = encoder_lstm(encoder_input)
    encoder_states = [state_h, state_c]  # Encoder state is passed to the decoder

    decoder_input = Input(shape=(forecast_steps, 1))  # Input for decoder
    decoder_lstm = LSTM(lstm_units, return_sequences=True, return_state=True, dropout=dropout_rate)
    decoder_output, _, _ = decoder_lstm(decoder_input, initial_state=encoder_states)
    decoder_dense = Dense(1)  # Output layer to make predictions
    decoder_output = decoder_dense(decoder_output)

    model = Model([encoder_input, decoder_input], decoder_output)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate), loss='mean_squared_error')
    return model

# Objective function for Optuna optimization
def objective(trial):
    # Suggest values for hyperparameters
    lstm_units = trial.suggest_int('lstm_units', 32, 128)  # Number of LSTM units
    dropout_rate = trial.suggest_uniform('dropout_rate', 0.0, 0.5)  # Dropout rate
    learning_rate = trial.suggest_loguniform('learning_rate', 1e-5, 1e-2)  # Learning rate range

    # Create the Seq2Seq model with these hyperparameters
    model = create_seq2seq_model(time_step=30, forecast_steps=5, lstm_units=lstm_units, 
                                 dropout_rate=dropout_rate, learning_rate=learning_rate)
    
    # Initialize decoder inputs for training and testing
    decoder_input_train = np.zeros((X_train_seq.shape[0], 5, 1))  # initialize with zeros for decoder input
    decoder_input_test = np.zeros((X_test_seq.shape[0], 5, 1))  # similarly for the test set
    
    # Fit the model
    model.fit([X_train_seq, decoder_input_train], y_train_seq, epochs=10, batch_size=64,
              validation_data=([X_test_seq, decoder_input_test], y_test_seq), verbose=0)
    
    # Prediction
    decoder_input_pred = np.zeros((X_test_seq.shape[0], 5, 1))  # initialize for predictions
    y_pred_seq2seq = model.predict([X_test_seq, decoder_input_pred], verbose=0)

    # We take the first value of each forecast sequence for comparison
    y_pred_seq2seq_first = y_pred_seq2seq[:, 0].flatten()
    y_test_first = y_test_seq[:, 0].flatten()

    # Calculating RMSE and MAE
    rmse = np.sqrt(mean_squared_error(y_test_first, y_pred_seq2seq_first))
    mae = mean_absolute_error(y_test_first, y_pred_seq2seq_first)

    # Return RMSE (it could be switched to MAE based on my preference)
    return rmse

# Running Optuna to optimize hyperparameters
study = optuna.create_study(direction='minimize')  # Minimize RMSE
study.optimize(objective, n_trials=50)

# Printing the best hyperparameters found
print(f'Best hyperparameters: {study.best_trial.params}')

# Getting the best hyperparameters from the optimization process
best_params = study.best_trial.params

# Creating the final Seq2Seq model using the optimized hyperparameters
best_model = create_seq2seq_model(time_step=30, forecast_steps=5,
                                  lstm_units=best_params['lstm_units'], 
                                  dropout_rate=best_params['dropout_rate'],
                                  learning_rate=best_params['learning_rate'])

# Retraining the model with the optimal parameters
decoder_input_train = np.zeros((X_train_seq.shape[0], 5, 1))  # initializing for training
decoder_input_test = np.zeros((X_test_seq.shape[0], 5, 1))  # similarly for test set

best_model.fit([X_train_seq, decoder_input_train], y_train_seq, epochs=20, batch_size=64,
               validation_data=([X_test_seq, decoder_input_test], y_test_seq))

# Predictions from the tuned model
decoder_input_pred = np.zeros((X_test_seq.shape[0], 5, 1))
y_pred_seq2seq = best_model.predict([X_test_seq, decoder_input_pred])

# Plotting actual vs predicted values
y_pred_seq2seq_first = y_pred_seq2seq[:, 0].flatten()
y_test_first = y_test_seq[:, 0].flatten()

test_dates = test_data['DateTime'].iloc[30 + 5:30 + 5 + len(y_test_first)].values

assert len(test_dates) == len(y_test_first) == len(y_pred_seq2seq_first), \
    f"Length mismatch: test_dates({len(test_dates)}), y_test({len(y_test_first)}), predictions({len(y_pred_seq2seq_first)})"

# Plot the results
plt.figure(figsize=(14, 6))
plt.plot(test_dates, y_test_first, color='blue', label='Actual Consumption')
plt.plot(test_dates, y_pred_seq2seq_first, color='red', label='Predicted Consumption')
plt.title('Actual vs Predicted Electricity Consumption (Seq2Seq - Optimized)')
plt.xlabel('DateTime')
plt.ylabel('Electricity Consumption (scaled)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Applying L2 regularization to the layers in our model. This will add a penalty for large weights, making the model generalize better.
from tensorflow.keras.regularizers import l2

def create_seq2seq_model(time_step, forecast_steps, lstm_units, dropout_rate, learning_rate):
    encoder_input = Input(shape=(time_step, 1))
    encoder_lstm = LSTM(lstm_units, return_state=True, dropout=dropout_rate,
                        kernel_regularizer=l2(1e-4))  # Add L2 regularization
    encoder_output, state_h, state_c = encoder_lstm(encoder_input)
    encoder_states = [state_h, state_c]

    decoder_input = Input(shape=(forecast_steps, 1))
    decoder_lstm = LSTM(lstm_units, return_sequences=True, return_state=True, dropout=dropout_rate,
                        kernel_regularizer=l2(1e-4))  # Add L2 regularization
    decoder_output, _, _ = decoder_lstm(decoder_input, initial_state=encoder_states)
    decoder_dense = Dense(1, kernel_regularizer=l2(1e-4))  # Add L2 regularization
    decoder_output = decoder_dense(decoder_output)

    model = Model([encoder_input, decoder_input], decoder_output)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate), loss='mean_squared_error')
    return model
```
